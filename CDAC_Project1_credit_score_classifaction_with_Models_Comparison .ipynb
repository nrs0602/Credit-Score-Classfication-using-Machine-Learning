```python
def calculate_months(value):
    if pd.isna(value):
        return np.nan
    parts = value.split(" and ")
    year = int(parts[0].split(" ")[0])
    month = int(parts[1].split(" ")[0])
    return (year * 12) + month
```

```python
trainData["Credit_History_Age"] = trainData["Credit_History_Age"].apply(calculate_months)
```

```python
def splitPaymentBehaviour(text):
    if pd.isna(text):

        return pd.Series([np.nan, np.nan])
    if "Low_spent" in text:
        spending = 0
    else:
        spending = 1

    if "Small_value" in text:
        value = 0
    elif "Medium_value" in text:
        value =1
    else:
        value = 2

    return pd.Series([spending,value])
```

```python
def cap_outliers(df,col_name):
    Q1 = df[col_name].quantile(0.25)
    Q3 = df[col_name].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    df.loc[df[col_name] < lower_limit, col_name] = lower_limit
    df.loc[df[col_name] > upper_limit, col_name] = upper_limit

    return df
```

```python
outlier_cols = [
    'Annual_Income',
    'Monthly_Inhand_Salary',
    'Outstanding_Debt',
    'Total_EMI_per_month',
    'Amount_invested_monthly',
    'Monthly_Balance'
]

for col in outlier_cols:
    trainData = cap_outliers(trainData,col)
    testData = cap_outliers(testData,col)
```


## Define Models for Comparison


```python
models = {
    "Logistic Regression": LogisticRegression(
        max_iter=500,
        n_jobs=-1
    ),

    "Decision Tree": DecisionTreeClassifier(
        random_state=42
    ),

    "Random Forest": RandomForestClassifier(
        n_estimators=150,
        random_state=42,
        n_jobs=-1
    ),

   # "SVM": SVC(
    #    kernel='rbf',
    #    gamma='scale',
     #   probability=False
   # ),

    "XGBoost": XGBClassifier(
        n_estimators=150,
        learning_rate=0.1,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        objective='multi:softmax',
        eval_metric='mlogloss',
        tree_method='hist',
        n_jobs=-1,
        random_state=42
    ),

    "LightGBM":LGBMClassifier(
        n_estimators=200,
        learning_rate=0.05,
        num_leaves=31,
        max_depth=-1,
        objective="multiclass",
        random_state=42,
        n_jobs=-1
    )
}
```

**Fast Cross-Validation** **Comparison**

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
import time
import pandas as pd

def compare_models_cv(models, X, y, folds=5):
    results = []
    skf = StratifiedKFold(n_splits=folds, shuffle=True, random_state=42)

    for name, model in models.items():
        start = time.time()

        scores = cross_val_score(
            model,
            X,
            y,
            cv=skf,
            scoring='accuracy',
            n_jobs=-1
        )

        end = time.time()

        results.append({
            "Model": name,
            "Mean Accuracy": scores.mean(),
            "Std Dev": scores.std(),
            "Time (sec)": round(end - start, 2)
        })

    return pd.DataFrame(results).sort_values(by="Mean Accuracy", ascending=False)
```

```python
results_df = compare_models_cv(models, X_train, y_train_encoded)
results_df
```

```python
best_model_name = results_df.iloc[0]["Model"]
best_model = models[best_model_name]

best_model.fit(X_train, y_train_encoded)
y_pred_encoded = best_model.predict(X_test)
```

```python
y_pred = label_encoder.inverse_transform(y_pred_encoded)

from sklearn.metrics import accuracy_score, classification_report

print(f"\nBest Model: {best_model_name}")
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
